```python

# Function
files1 = glob("../../data/raw/MetaMotion/*.csv")

def read_data_from_files(files):
    '''
    Read Raw date and convert into dataframe
    '''

    acs_df=pd.DataFrame()
    gyro_df=pd.DataFrame()

    acs_set=1
    gyro_set=1

    for f in files:

        fs=f.split("-")
        participents= fs[0].replace(data_path, "")
        label= fs[1]
        category= fs[2].rstrip("123").rstrip("_MetaWear_2019")
        
        df=pd.read_csv(f)

        df["participents"]=participents
        df["label"]=label
        df["category"]=category

        if "Accelerometer" in f:
            df["set"]=acs_set
            acs_set+=1
            acs_df=pd.concat([acs_df,df])

        if "Gyroscope" in f:
            df["set"]=gyro_set
            gyro_set+=1
            gyro_df=pd.concat([gyro_df,df])
        

    acs_df.index=pd.to_datetime(acs_df["epoch (ms)"],unit="ms")
    gyro_df.index=pd.to_datetime(gyro_df["epoch (ms)"],unit="ms")


    del acs_df["epoch (ms)"]
    del acs_df["time (01:00)"]
    del acs_df["elapsed (s)"]

    del gyro_df["epoch (ms)"]
    del gyro_df["time (01:00)"]
    del gyro_df["elapsed (s)"]

    return acs_df,gyro_df
```

```python
data_marged=pd.concat([acs_df.iloc[:,:3],gyro_df],axis=1)
```

```python
data_marged.columns=[
        "Accelerometer_x",
        "Accelerometer_y",
        "Accelerometer_z",
        "Gyroscope_x",
        "Gyroscope_y",
        "Gyroscope_z",
        "Participents",
        "Label",
        "Category",
        "Set"
]


sampling={
        "Accelerometer_x":"mean",
        "Accelerometer_y":"mean",
        "Accelerometer_z":"mean",
        "Gyroscope_x":"mean",
        "Gyroscope_y":"mean",
        "Gyroscope_z":"mean",
        "Participents":"last",
        "Label":"last",
        "Category":"last",
        "Set":"last"
}


data_marged[:1000].resample(rule="200ms").apply(sampling)
```

```python
days=[g for a,g in data_marged.groupby(pd.Grouper(freq="D"))]
data_resampled=pd.concat([df.resample(rule="200ms").apply(sampling).dropna() for df in days])
data_resampled["Set"]=data_resampled["Set"].astype("int")
```
